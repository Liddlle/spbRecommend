```python
import urllib.request
import urllib.parse
import simplejson as json
import pandas as pd
import re
```

```python
events_list = []
events_list_dates = []
url = 'https://kudago.com/public-api/v1.3/events/?location=spb&page_size=100&fields=id,publication_date,dates,title,slug,place,description,body_text,categories,tagline,age_restriction,price,is_free,images,favorites_count,tags,site_url,images&expand=date'
```

```python
def parse_list(url):
    user_agent = 'Mozilla/5.0 (Windows NT 6.1; Win64; x64)'
    headers = {'User-Agent': user_agent}
    req = urllib.request.Request(url, headers=headers)
    try:
        with urllib.request.urlopen(req) as response:
            the_page = response.read()
    except urllib.error.URLError as e:
        print(e.reason)

    data = json.loads(the_page)
    return data

def clear(text):
    cleared = re.sub(r"[\r\n\t]+", " ", str(text))
    return cleared
```

```python
obj = parse_list(url)
cnt = 0
```

```python
while True:
    if obj['results']:
        tmp_list = []
        tmp_list_d = []
        cnt += 1
        print(cnt)
        for i in range(len(obj['results'])):
            loc = []
            loc.append(obj['results'][i]['id'])
            loc.append(obj['results'][i]['publication_date'])
            for d in range(len(obj['results'][i]['dates'])):
                loc_d = []
                loc_d.append(obj['results'][i]['id'])
                loc_d.append(obj['results'][i]['dates'][d]['start'])
                loc_d.append(obj['results'][i]['dates'][d]['end'])
                tmp_list_d.append(loc_d)
            loc.append(obj['results'][i]['title'])
            loc.append(obj['results'][i]['slug'])
            place = obj['results'][i]['place']
            loc.append('' if place is None else place['id'])
            loc.append(clear(obj['results'][i]['body_text']))
            loc.append(clear(obj['results'][i]['description']))
            loc.append(obj['results'][i]['site_url'])
            loc.append(obj['results'][i]['tagline'])
            loc.append(obj['results'][i]['age_restriction'])
            loc.append(obj['results'][i]['price'])
            loc.append(obj['results'][i]['is_free'])
            if obj['results'][i]['images']:
                loc.append(obj['results'][i]['images'][0]['image'])
            else:
                loc.append('')
            loc.append(obj['results'][i]['favorites_count'])
            categories = '###'.join(obj['results'][i]['categories'])
            loc.append(categories)
            tags = '###'.join(obj['results'][i]['tags'])
            loc.append(tags)
            tmp_list.append(loc)
        events_list = events_list + tmp_list
        events_list_dates = events_list_dates + tmp_list_d
        if obj['next']:
            next1 = obj['next']
            obj = []
            obj = parse_list(next1)
        else:
            break
```

```python
events_table = pd.DataFrame(events_list, columns=['id','publication_date','title','slug','place_id',
                                                  'body_text','description','site_url','tagline','age_restriction',
                                                  'price','is_free','image','favorites_count','categories','tags'])
```
